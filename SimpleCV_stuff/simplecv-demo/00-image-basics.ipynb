## Digression: colour spaces

### RGB colour space

```python
Image('media/rgb-cube.png')
```

```python
img = Image('media/rainbow-cake.jpg').scale(0.3)
img
```

```python
img_s = img.scale(0.5)
three = img_s.sideBySide(img_s).sideBySide(img_s)

r, g, b = img_s.splitChannels(True)
channels = r.sideBySide(g).sideBySide(b)

three.sideBySide(channels, side='bottom')
```

### HSV color space (Hue, Saturation, Value)

```python
Image('media/hsv-cone.png')
```

```python
img_s = img.scale(0.5)
three = img_s.sideBySide(img_s).sideBySide(img_s)

v, s, h = img_s.toHSV().splitChannels()
channels = h.sideBySide(s).sideBySide(v)

three.sideBySide(channels, side='bottom')
```

## Operations on images

```python
img = Image('media/kitten.jpg').scale(0.4)
img
```

### Cropping

```python
img[100:300, 50:200]
```

```python
img.crop(100, 50, 200, 150)
```

### Rotating

```python
img.rotate(90)
```

```python
img.rotate(90, fixed=False)
```

```python
img.rotate(45)
```

### Flipping

```python
img.flipHorizontal()
```

```python
img.flipVertical()
```

### Blitting

Think copy-paste for images

```python
smallkitty = img.scale(0.3)
img.blit(smallkitty, (30, 20))
```

### Side by side

```python
img.sideBySide(img.invert(), 'right')
```
